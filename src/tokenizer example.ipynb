```python
from transformers import AutoTokenizer
```

```python
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
# Shows you how a given text is broken into tokens. Ignore the Ġ characters. This is just a reserved special 
# character for how the tokenizer is representing spaces
text = 'I drank a coca-cola.'
print(tokenizer.tokenize(text))
```

```python
# returns dictionary of all tokens, mapped to the integer id that gets fed into the LLM
tokenizer.get_vocab()
```
